```python
numa = 100
xt = np.zeros(numa)
xt[:10] = 3.1 # making first 10 terms equalt to 3.1
xt[-9:] = 3.1# making last 10 terms equal to 3.1
#this configuration let us achieve some kind of a sqaure signal
plt.plot(xt, color = "purple")
plt.title("plot of x(t)")
plt.xlabel("Sample index")
plt.ylabel("value of x(t)")
```

## What we should we carefull about most is that FT will return us a result that has both real and imaginary parts trying to print them together may or may not cause problems depending on context

```python
fxt = np.fft.fft(xt)
plt.subplot(1, 2, 1)
#First Plot, Real part of the FT
plt.plot(np.real(fxt), color = "yellow")
plt.title("Plot of real part of FT of x(t)")
plt.ylabel("Magnitude")
plt.xlabel("Frequency Index")

plt.subplot(1, 2, 2)
#Second Plot, Imaginary part of the FT
plt.plot(np.imag(fxt), color = "yellow")
plt.title("Plot of imaginary part of FT of x(t)")
plt.ylabel("Magnitude")
plt.xlabel("Frequency Index")
```

## As you see above since square function is even imaginary part of the fft is 0 after ignoring the numerical precision

## But something is wrong, this is not the sinc function we expext from a sqaure signal

```python
fxt = np.fft.fft(xt)
plt.subplot(1, 2, 1)
#First Plot, Real part of the FT
plt.plot(np.fft.fftshift(np.real(fxt)), color = "green")
plt.title("Plot of real part of FT of x(t)")
plt.ylabel("Magnitude")
plt.xlabel("Frequency Index")

plt.subplot(1, 2, 2)
#Second Plot, Imaginary part of the FT
plt.plot(np.fft.fftshift(np.imag(fxt)), color = "green")
plt.title("Plot of imaginary part of FT of x(t)")
plt.ylabel("Magnitude")
plt.xlabel("Frequency Index")
```
